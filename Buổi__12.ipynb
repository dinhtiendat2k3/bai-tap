```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the dataset
file_path = '/content/penguins.csv'
penguins_data = pd.read_csv(file_path)

# Drop rows with missing values for the relevant columns
penguins_data_clean = penguins_data.dropna(subset=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm'])

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Define the axes
x = penguins_data_clean['bill_length_mm']
y = penguins_data_clean['bill_depth_mm']
z = penguins_data_clean['flipper_length_mm']
c = penguins_data_clean['species']

# Create the scatter plot
sc = ax.scatter(x, y, z, c=pd.factorize(c)[0], cmap='viridis', label=c)

# Add labels
ax.set_xlabel('Bill Length (mm)')
ax.set_ylabel('Bill Depth (mm)')
ax.set_zlabel('Flipper Length (mm)')
plt.title('3D Scatter Plot of Penguin Measurements')

# Create a legend
handles, labels = sc.legend_elements()
legend_labels = penguins_data_clean['species'].unique()
ax.legend(handles, legend_labels, loc='upper left')

# Show plot
plt.show()
```
