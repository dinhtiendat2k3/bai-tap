```python
# Lọc dữ liệu cho các quốc gia cần vẽ biểu đồ (Hoa Kỳ, Tây Ban Nha, Trung Quốc, Ấn Độ) và giai đoạn từ 1961 đến 2020
countries = ['United States', 'Spain', 'China', 'India']
filtered_data = data[(data['Entity'].isin(countries)) & (data['Year'] >= 1961) & (data['Year'] <= 2020)]

# Tạo DataFrame mới để lưu trữ dữ liệu tổng hợp từ các quốc gia
plot_data = pd.DataFrame(columns=['Year', 'Average Yield'])

# Tính trung bình năng suất của các quốc gia và thêm vào DataFrame mới
for year in range(1961, 2021):
    yearly_data = filtered_data[filtered_data['Year'] == year]
    average_yield = yearly_data['Tomatoes | 00000388 || Yield | 005419 || tonnes per hectare'].mean()
    plot_data.loc[len(plot_data)] = [year, average_yield]

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.plot(plot_data['Year'], plot_data['Average Yield'])
plt.title('Năng suất cà chua trung bình (tấn/ha) từ 1961 đến 2020')
plt.xlabel('Năm')
plt.ylabel('Năng suất trung bình (tấn/ha)')
plt.grid(True)
plt.show()
```
